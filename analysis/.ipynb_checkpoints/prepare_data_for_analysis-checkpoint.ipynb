```python
import pandas as pd
import os

file = "prediction_audio_noise.csv"
out_file = "prepared_data_audio_noise.csv"

path = os.path.join("..", "out", file)
predictions = pd.read_csv(path)

out_path = os.path.join("out", out_file)

splits = predictions['split'].unique()
split_data = {}

for split in splits:
    split_data[split] = predictions[predictions['split'] == split]

classes = ['positive', 'negative', 'neutral']
metrics = pd.DataFrame(columns=['split', 'class', 'TP', 'FP', 'FN', 'TN'])

for split, data in split_data.items():
    for cl in classes:
        tp = data[data['prediction'] == cl and data['label'] == cl]
        fp = data[data['prediction'] == cl and data['label'] != cl]
        fn = data[data['prediction'] != cl and data['label'] == cl]
        tn = data[data['prediction'] != cl and data['label'] != cl]
        metrics.loc[len(metrics)] = [split, cl, len(tp), len(fp), len(fn), len(tn)]

metrics.to_csv(out_path, index=False)
```
